```python
import altair as alt
import numpy as np
import pandas as pd
```

```python
cust_demo_add_df = pd.read_excel('cleaned.xlsx',header=0, sheet_name='CustomerDemographic+Address')
cust_demo_add_df
```

```python
all_df = pd.read_excel('cleaned.xlsx',header=0, sheet_name='All')
all_df
```

```python
top_1000 = all_df.groupby(by=['customer_id','first_name','last_name','address','postcode','state','country'],as_index=False)['Profit'].sum().sort_values(by='Profit',ascending=False).head(1000).reset_index().drop(columns='index')
writer = pd.ExcelWriter('/Users/Sunny/KPMG Data Analytics Virtual Experience Program/Phase 3 Data Insights and Presentation/top1000.xlsx', engine='xlsxwriter')
top_1000.to_excel(writer, sheet_name='Top 1000 Customers', index=False)
writer.save()
top_1000
```

```python
profit_state_industry = alt.Chart(all_df.groupby(by=['state','job_industry_category'],as_index=False)['Profit'].sum(),title='Total Profit by State and Industry').mark_bar().encode(
    x=alt.X('job_industry_category',title='Job Industry Category'),
    y=alt.Y('Profit'),
    color='job_industry_category',
    column=alt.Column('state', title='State')
)
profit_state_industry
```

```python
monthly = all_df.set_index('transaction_date').groupby(pd.Grouper(freq='M'))['Profit'].sum().reset_index()
monthly_profit = alt.Chart(monthly,title='Monthly Total Profit').mark_line().encode(
    x=alt.X('month(transaction_date)',title='Month'),
    y=alt.Y('Profit',title='Total Profit')
)
monthly_profit
```

```python
alt.Chart(all_df.groupby(by=['gender','AGE'],as_index=False)['Profit'].sum(), title='Total Profit with Age and Gender').mark_bar().encode(
    x=alt.X('AGE',bin=True,title='AGE'),
    y=alt.Y('sum(Profit)',title='Total Profit'),
    color=alt.Color('gender',title='Gender')
)
```

```python
alt.Chart(all_df.groupby(by=['wealth_segment','job_industry_category'],as_index=False)['Profit'].sum(),title='Total Profit by Job Industry and Wealth Segment').mark_bar().encode(
    x=alt.X('job_industry_category',title='Job Industry Category'),
    y=alt.Y('Profit', title='Total Profit'),
    color='job_industry_category',
    column=alt.Column('wealth_segment', title='Wealth Segment')
)
```

```python
alt.Chart(all_df.groupby(by=['wealth_segment','product_class'],as_index=False)['Profit'].sum().reindex([1,2,0,4,5,3,7,8,6]), title='Total Profit with Wealth Segment and Product Class').mark_bar().encode(
    x=alt.X('product_class',title='Product Class',sort=None),
    y=alt.Y('sum(Profit)',title='Total Profit'),
    color=alt.Color('wealth_segment',title='Wealth Segment')
)
```

```python
alt.Chart(all_df.groupby(by=['wealth_segment','brand'],as_index=False)['Profit'].sum(),title='Total Profit by Wealth Segment and Brand').mark_bar().encode(
    x=alt.X('brand',title='Brand'),
    y=alt.Y('Profit', title='Total Profit'),
    color=alt.Color('wealth_segment',title='Wealth Segment'),
    column=alt.Column('wealth_segment', title='Wealth Segment')
)
```
